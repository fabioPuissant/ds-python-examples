**Trying to read the data from the supermarkets data-frame and setting it in the format (street, city, state, country) so geopy can get the LatLong of that adderess**

```python
#First we try to get all these variables in a string and save them in a column in the data-frame
df["Format_Address"] = df["Address"] + "," + df["City"] + "," + df["State"] + "," + df["Country"]
df

#Can also save it in a list
#list( df["Address"] + "," + df["City"] + "," + df["State"] + "," + df["Country"])
```

```python
# With the apply method, we can apply a function to a selection of a data-frame!

# In this case we selected the  each data-cell of the column 'Format_Address' and give this data to the nom.geocode() method
#   to calculate the LatLong of each address in the supermarkets data-frame

#After calculating the LatLongs, the data is stored in a new column named "Coordinates in our supermarkets data-frame!
df["Coordinates"] = df["Format_Address"].apply(nom.geocode)
df
```

```python
#Example of what is stored in Coordinates (the first in this selection -> [0]!)
df["Coordinates"][0]
df.Coordinates[0]
```

```python
#Since the data-type of the stored values in the Coordinates column is of type 'Location', 
#  we can use the Location-type methods!
# See all possible methods ==> dir(df.Coordinates[0])
df.Coordinates[0].latitude
```

```python
df.Coordinates[0].longitude
```

```python
# Now it is possible that we want to have the Longtitude and Latitude of all supermarkets in our data-frame,
# in separate column 
#The next will give an error cause your getting back a selection and not a Location-type object!
#df["Coordinates"].latitude

# SOLUTION: ==> use the apply() method combined with a lambda!
df["Latitude"] = df["Coordinates"].apply(lambda x: x.latitude if x != None else None)
df["Longitude"] = df["Coordinates"].apply(lambda x: x.longitude if x != None else None)
df
```
